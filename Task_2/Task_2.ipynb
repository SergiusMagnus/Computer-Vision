```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv.imread("data/logs-1.jpg")
src = cv.cvtColor(src, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(src, cv.COLOR_RGB2GRAY)
gray = cv.blur(gray, (3, 3))
detected_circles = cv.HoughCircles(gray,
                   cv.HOUGH_GRADIENT,
                   2,  # Inverse ratio of the accumulator resolution to the image resolution
                   10, # Minimum distance between the centers of the detected circles
                   param1 = 200, # The higher threshold of the two passed to the Canny edge detector
                   param2 = 30,  # The accumulator threshold for the circle centers at the detection stage
                   minRadius = 5, maxRadius = 15)
image = src.copy()
logs_count = 0

if detected_circles is not None:
    detected_circles = np.uint16(np.around(detected_circles))

    for points in detected_circles[0, :]:
        a, b, r = points[0], points[1], points[2]
        cv.circle(image, (a, b), r, (0, 255, 0), 2)
        cv.circle(image, (a, b), 1, (255, 0, 0), 3)
        logs_count += 1

print("Total logs: ", logs_count)
plt.imshow(image)
```
